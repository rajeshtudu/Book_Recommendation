```python
# looking the total no of ratings given by the users
ratings['user_id'].value_counts()
```

```python
# changing the ratings dataframe into new dataframe who has given more than 200 ratings
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
# mergings the books with the ratings
ratings_with_books = ratings.merge(books, on = 'ISBN')
```

```python
# looking at dataframe
ratings_with_books.head()
```

```python
# looking at its shape
ratings_with_books.shape
```

```python
# looking at the books with the no of ratings
num_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_ratings.head()
```

```python
# changing the name of rating
num_ratings.rename(columns={'rating': 'no_of_ratings'}, inplace = True)
```

```python
# looking at its head
num_ratings.head()
```

```python
# creating final rating dataframe
final_rating = ratings_with_books.merge(num_ratings, on = 'title')
```

```python
# looking at its head
final_rating.head()
```

```python
# looking at its shape
final_rating.shape
```

```python
# excluding the ratings of the books less than 50
final_rating = final_rating[final_rating['no_of_ratings']>=50]
```

```python
final_rating.head()
```

```python
# looking at the sample dataframe of final rating
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
# dropping the duplicate user_id and title
final_rating.drop_duplicates(['user_id', 'title'], inplace = True)
```

```python
final_rating.shape
```

```python
# creating a pivot table
book_pivot = final_rating.pivot_table(columns = 'user_id', index = 'title', values = 'rating')
```

```python
# looking at its head
book_pivot.head()
```

```python
# looking at its shape
book_pivot.shape
```

```python
# filling in the missing values with 0
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

### Modeling

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors = 6)
```

```python
distance
```

```python
suggestion
```

```python
# creating a for loop to test out the recommendation
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
# saving the index in books_name
books_name = book_pivot.index
```

```python
# creating a function to recommend a book
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors = 6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
# testing recommendation system
book_name = 'Zoya'
recommend_book(book_name)
```
